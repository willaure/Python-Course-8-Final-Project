```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Filter for recession periods
df_rec = df[df['Recession'] == 1]

# Calculate average sales by year and vehicle type during recession
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Calculate normalized sales (divided by mean sales for each vehicle type)
df_Mline['Normalized_Sales'] = df_Mline.groupby('Vehicle_Type')['Automobile_Sales'].transform(lambda x: x / x.mean())

# Set Year as index
df_Mline.set_index('Year', inplace=True)

# Create plot
plt.figure(figsize=(12, 8))
for vehicle_type in df_Mline['Vehicle_Type'].unique():
    data = df_Mline[df_Mline['Vehicle_Type'] == vehicle_type]
    plt.plot(data.index, data['Normalized_Sales'],
             label=vehicle_type,
             marker='o',
             linewidth=2)

# Highlight recession years
recession_years = df_rec['Year'].unique()
for year in recession_years:
    plt.axvline(x=year, color='red', linestyle='--', alpha=0.3)

# Add labels and title
plt.legend(title="Vehicle Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Normalized Sales (Mean = 1.0)", fontsize=12)
plt.xlabel("Year", fontsize=12)
plt.title("Normalized Automobile Sales by Vehicle Type During Recession Periods",
          fontsize=14, pad=20)

# Add grid and adjust layout
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()

# Additional analysis
print("\nKey Insights:")
print("1. Values above 1.0 indicate better-than-average performance for that vehicle type")
print("2. Values below 1.0 indicate worse-than-average performance")
print("3. The red dashed lines mark all recession years in the dataset")
print("4. Comparing slopes shows which vehicle types recover fastest post-recession")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load and prepare data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.year

# Create comparison dataframe
df_compare = df.groupby(['Year', 'Vehicle_Type', 'Recession'])['Automobile_Sales'].mean().reset_index()
df_compare['Recession_Period'] = df_compare['Recession'].map({0: 'Non-Recession', 1: 'Recession'})

# Set style
sns.set(style="whitegrid", palette="pastel", rc={"figure.figsize":(14, 8)})

# Create facet grid
g = sns.FacetGrid(df_compare, col="Vehicle_Type", hue="Recession_Period",
                  col_wrap=3, height=4, aspect=1.5, sharey=False)
g.map(sns.lineplot, "Year", "Automobile_Sales", marker="o", linewidth=2.5)
g.set_titles("{col_name}")

# Add recession highlights and styling
for ax in g.axes.flat:
    recession_years = df_compare[(df_compare['Recession'] == 1) &
                               (df_compare['Vehicle_Type'] == ax.get_title())]['Year']
    for year in recession_years:
        ax.axvspan(year-0.5, year+0.5, color='red', alpha=0.1)
    ax.set_ylabel('Average Sales')
    ax.set_xlabel('Year')
    ax.set_xticks(range(1980, 2024, 5))

g.add_legend(title="Economic Period")
plt.suptitle("Vehicle Sales Trends: Recession vs. Non-Recession Periods", y=1.03, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load and prepare data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)
df['Recession_Period'] = df['Recession'].map({1: 'Recession', 0: 'Non-Recession'})

# Set visual style
sns.set_style("whitegrid", {'grid.linestyle': '--', 'axes.edgecolor': '0.3'})
plt.figure(figsize=(14, 8))
palette = {"Recession": "#FF6B6B", "Non-Recession": "#4ECDC4"}

# Create violin plot with swarm overlay
ax = sns.violinplot(
    x='Vehicle_Type',
    y='Automobile_Sales',
    hue='Recession_Period',
    data=df,
    split=True,
    palette=palette,
    inner="quartile",
    linewidth=1.5
)

# Add swarm plot for data point distribution
sns.swarmplot(
    x='Vehicle_Type',
    y='Automobile_Sales',
    hue='Recession_Period',
    data=df,
    dodge=True,
    palette=palette,
    size=3,
    alpha=0.4,
    ax=ax
)

# Customize plot
plt.title('Automobile Sales Distribution: Recession vs. Non-Recession Periods', pad=20, fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Sales Volume (Units)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Economic Condition', bbox_to_anchor=(1, 1))

# Add statistical annotations
for i, vehicle_type in enumerate(df['Vehicle_Type'].unique()):
    rec_data = df[(df['Vehicle_Type'] == vehicle_type) & (df['Recession'] == 1)]['Automobile_Sales']
    non_rec_data = df[(df['Vehicle_Type'] == vehicle_type) & (df['Recession'] == 0)]['Automobile_Sales']
    diff = non_rec_data.mean() - rec_data.mean()
    ax.text(i, ax.get_ylim()[1]*0.95, f"Δ{diff:.0f}", ha='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
# Create dataframe for recession period
rec_data = df[df['Recession'] == 1]

# Create scatter plot with improved styling
plt.figure(figsize=(10, 6))
plt.scatter(
    x=rec_data['Price'],  # Assuming 'Price' column exists - if not, use 'Vehicle_Price'
    y=rec_data['Automobile_Sales'],
    c='crimson',          # Recession-themed color
    alpha=0.6,            # Transparency for overlapping points
    edgecolor='black',     # Point borders
    linewidth=0.5
)

# Add regression line to show trend
import numpy as np
z = np.polyfit(rec_data['Price'], rec_data['Automobile_Sales'], 1)
p = np.poly1d(z)
plt.plot(rec_data['Price'], p(rec_data['Price']), "k--", linewidth=1.5)

# Add labels and title
plt.xlabel('Vehicle Price ($)', fontsize=12)
plt.ylabel('Sales Volume', fontsize=12)
plt.title('Vehicle Price vs Sales During Recessions', fontsize=14, pad=15)

# Add grid and correlation annotation
plt.grid(True, linestyle='--', alpha=0.3)
corr = rec_data['Price'].corr(rec_data['Automobile_Sales'])
plt.text(0.05, 0.95, f'Correlation: {corr:.2f}',
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for recession period and calculate ad expenditure by vehicle type
rec_data = df[df['Recession'] == 1]
ad_by_vehicle = rec_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().sort_values(ascending=False)

# Create the pie chart
plt.figure(figsize=(12, 8))

# Customize colors and plot parameters
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974']  # Professional color palette
explode = [0.05] * len(ad_by_vehicle)  # Slightly separate all slices

# Generate the pie chart
wedges, texts, autotexts = plt.pie(
    ad_by_vehicle,
    labels=ad_by_vehicle.index,
    colors=colors,
    autopct=lambda p: f'{p:.1f}%',
    startangle=140,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1},
    textprops={'fontsize': 12}
)

# Add percentage labels inside slices
plt.setp(autotexts, color='white', weight='bold')

# Add title and legend
plt.title('Advertisement Expenditure by Vehicle Type During Recessions',
          fontsize=16, pad=20, weight='bold')

# Create legend with absolute values
legend_labels = [f'{label}: ${value/1e6:.2f}M'
                 for label, value in zip(ad_by_vehicle.index, ad_by_vehicle)]
plt.legend(wedges, legend_labels,
           title='Total Expenditure',
           loc='center left',
           bbox_to_anchor=(1, 0.5))

# Equal aspect ratio ensures pie is circular
plt.axis('equal')
plt.tight_layout()
plt.show()
```
